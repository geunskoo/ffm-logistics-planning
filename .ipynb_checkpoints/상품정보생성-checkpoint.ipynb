```python
import pandas as pd
import numpy as np

df=pd.read_excel('/Users/yoon/Desktop/agv/202204_출고_주문번호추가.xlsx',header=1)
df=df.rename(columns={'상품명':'상품'})
df1=df[["상품코드","상품"]]

df2=pd.DataFrame({"상품":df1["상품"].value_counts().index,"개수":df1["상품"].value_counts().values})

df3=pd.merge(df1,df2,on="상품")

df4=df3.drop_duplicates()
df4
```

```python
df5 = pd.DataFrame({
    '상품':df4["상품"],
    'ABC': [['A', 'B', 'C'][np.random.randint(0, 3)] for i in range(len(df4))], 
    '단품가로(mm)':  [np.random.randint(1,100) for i in range(len(df4))], 
    '단품세로(mm)': [np.random.randint(1,100) for i in range(len(df4))], 
    '단품높이(mm)':  [np.random.randint(1,100) for i in range(len(df4))], 
    '중량(g)':  [np.random.randint(1, 300) for i in range(len(df4))]
})
df5
```

```python
df6=pd.merge(df4,df5)

df7=df6[["상품코드","상품","ABC","단품가로(mm)","단품세로(mm)","단품높이(mm)","중량(g)","개수"]]
```

```python
df7.to_excel('productInformation.xlsx')
```
