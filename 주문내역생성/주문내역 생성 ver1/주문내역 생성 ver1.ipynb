# 5. 1~4 내용 바탕으로 실전 요구사항에 더 적합한 프로그램 만들기.

```python
#특성개수
#len(col) ------------------------------------------------------------ 변수 1 (default - 32개(적재 데이터 기준))
#특성 생성하는 함수가 필요할 수도 있음. input - 특성 수/txt 파일 output = txt파일
col = np.loadtxt('C:/geunskoo-github-blog/ffm-logistics-planning/주문내역생성 ver1/특성.txt', delimiter=',', dtype=str, encoding='utf-8')
print(col)
```

```python
row = 10000 #default ------------------------------------------------ 변수 2 (주문개수)
#Black Box를 수정하면 주문 생성을 여러개 할 수 있음.

#name tag
alphabet = []
for i in string.ascii_uppercase:
    for j in string.ascii_uppercase:
        alphabet.append(i+j)
        
#name tag + randomint(0,10000)
data = {}
for i in range(len(col)):
    serial_num = []
    for j in range(row):
        serial_num.append(alphabet[i] + str(random.randint(0,10000)))
    data[col[i]] = serial_num
 
#DataFrame변환
order_list2 = pd.DataFrame(data,columns=col)
order_list2
```

```python
#엑셀 파일로 내보내기. -----------------------------------> 변수 3 주문 내역 Xlsx 파일이름. (default - 주문내역)
file_name ='주문내역.xlsx'
order_list2.to_excel(file_name)
print('ORDER LIST ---  successfully exported into Excel File')
```
